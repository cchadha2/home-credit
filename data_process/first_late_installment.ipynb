```python
import os
import math
import gc

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

%matplotlib inline
```

```python
prev = pd.read_csv('../data/previous_application.csv')
cc = pd.read_csv('../data/credit_card_balance.csv')
ins = pd.read_csv('../data/installments_payments.csv')
pos = pd.read_csv('../data/POS_CASH_balance.csv')
```

```python
cc_first_late = cc[['SK_ID_PREV', 'SK_ID_CURR', 'MONTHS_BALANCE', 'SK_DPD', 'SK_DPD_DEF']]
cc_first_late.sort_values(['SK_ID_PREV', 'SK_ID_CURR', 'MONTHS_BALANCE', 'SK_DPD', 'SK_DPD_DEF'])
```

```python
cc_first_late[(cc_first_late['SK_DPD'] > 0)  | (cc_first_late['SK_DPD_DEF'] > 0) ].sort_values(['SK_ID_PREV', 'MONTHS_BALANCE'])
```

```python
cc_first_late_month = cc_first_late[(cc_first_late['SK_DPD'] > 0)  | (cc_first_late['SK_DPD_DEF'] > 0) ].sort_values(['SK_ID_PREV', 'MONTHS_BALANCE']).groupby(['SK_ID_PREV', 'SK_ID_CURR']).agg({'MONTHS_BALANCE': ['min']})
cc_first_late_month_large = cc_first_late[cc_first_late['SK_DPD_DEF'] > 0].sort_values(['SK_ID_PREV', 'MONTHS_BALANCE']).groupby(['SK_ID_PREV', 'SK_ID_CURR']).agg({'MONTHS_BALANCE': ['min']})

```

```python
cc_first_late_month.columns = pd.Index(['CC_FIRST_MONTH_LATE'])
cc_first_late_month_large.columns = pd.Index(['CC_FIRST_MONTH_LATE_DEF'])
```

```python
cc_first_late_month
```

```python
cc_first_late_month_large
```

```python
prev_days = prev[['SK_ID_CURR', 'SK_ID_PREV', 'DAYS_FIRST_DUE', 'DAYS_FIRST_DRAWING',
      'DAYS_LAST_DUE_1ST_VERSION','DAYS_LAST_DUE', 'DAYS_TERMINATION']].sort_values(['SK_ID_CURR', 'SK_ID_PREV'])
```

```python
df = prev_days.merge(cc_first_late_month, on=['SK_ID_PREV', 'SK_ID_CURR'], how = 'left')
df = df.merge(cc_first_late_month_large, on=['SK_ID_PREV', 'SK_ID_CURR'], how = 'left')
```

```python
df
```

```python
df['CC_FIRST_LATE_INSTALMENT'] = df['CC_FIRST_MONTH_LATE']-(df['DAYS_FIRST_DUE']/30).round()
df['CC_FIRST_LATE_INSTALMENT_DRAWING'] = df['CC_FIRST_MONTH_LATE']-(df['DAYS_FIRST_DRAWING']/30).round()
df['CC_FIRST_LATE_INSTALMENT_DEF'] = df['CC_FIRST_MONTH_LATE_DEF']-(df['DAYS_FIRST_DUE']/30).round()
df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF'] = df['CC_FIRST_MONTH_LATE_DEF']-(df['DAYS_FIRST_DRAWING']/30).round()
```

```python
df['CC_LOAN_TERM_INSTALMENT'] = (df['DAYS_LAST_DUE']-df['DAYS_FIRST_DUE'])/30
df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION'] = (df['DAYS_LAST_DUE_1ST_VERSION']-df['DAYS_FIRST_DUE'])/30
df['CC_LOAN_TERM_INSTALMENT_TERMINATION'] = (df['DAYS_TERMINATION']-df['DAYS_FIRST_DUE'])/30
df['CC_LOAN_TERM_INSTALMENT_DRAWING'] = (df['DAYS_LAST_DUE']-df['DAYS_FIRST_DRAWING'])/30
df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION_DRAWING'] = (df['DAYS_LAST_DUE_1ST_VERSION']-df['DAYS_FIRST_DRAWING'])/30
df['CC_LOAN_TERM_INSTALMENT_TERMINATION_DRAWING'] = (df['DAYS_TERMINATION']-df['DAYS_FIRST_DRAWING'])/30
```

```python
df['CC_FIRST_LATE_IN_LOAN_TERM'] = (df['CC_FIRST_LATE_INSTALMENT']*100)/df['CC_LOAN_TERM_INSTALMENT']
df['CC_FIRST_LATE_IN_LOAN_TERM_1ST_VERSION'] = (df['CC_FIRST_LATE_INSTALMENT']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION']
df['CC_FIRST_LATE_IN_LOAN_TERM_TERMINATION'] = (df['CC_FIRST_LATE_INSTALMENT']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING'] = (df['CC_FIRST_LATE_INSTALMENT']*100)/df['CC_LOAN_TERM_INSTALMENT_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_1ST_VERSION_DRAWING'] = (df['CC_FIRST_LATE_INSTALMENT']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_TERMINATION_DRAWING'] = (df['CC_FIRST_LATE_INSTALMENT']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION_DRAWING']

df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_2'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING']*100)/df['CC_LOAN_TERM_INSTALMENT']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_1ST_VERSION'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_TERMINATION'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_DRAWING'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING']*100)/df['CC_LOAN_TERM_INSTALMENT_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_1ST_VERSION_DRAWING'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_TERMINATION_DRAWING'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION_DRAWING']
```

```python
df['CC_FIRST_LATE_IN_LOAN_TERM_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT']
df['CC_FIRST_LATE_IN_LOAN_TERM_1ST_VERSION_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION']
df['CC_FIRST_LATE_IN_LOAN_TERM_TERMINATION_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_1ST_VERSION_DRAWING_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_TERMINATION_DRAWING_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION_DRAWING']

df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_2_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_1ST_VERSION_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_TERMINATION_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_DRAWING_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_1ST_VERSION_DRAWING_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_1ST_VERSION_DRAWING']
df['CC_FIRST_LATE_IN_LOAN_TERM_DRAWING_TERMINATION_DRAWING_DEF'] = (df['CC_FIRST_LATE_INSTALMENT_DRAWING_DEF']*100)/df['CC_LOAN_TERM_INSTALMENT_TERMINATION_DRAWING']
```
